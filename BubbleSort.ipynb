```python
a = [3,2,5,3,67,8,45,2,4,34,4]

n = len(a)

c = 0

for i in range(n):
    for j in range(n-1):
        c=c+1

        if a[j] > a[j+1]:
            t=a[j]
            a[j]=a[j+1]
            a[j]=t

print("-------OUTPUT-------")

for i in a:
    print(i,end=" ")

print("Total Hits : " , c)


    
```

```python
# OPTIMISED CODE 
# Adjacent swapping / neighbour swapping

a = [5,4,3,4,5,2,1]

n = len(a)

c=0

for i in range(n):
    f=False
    for j in range(n-i-1):
        c=c+1
        if a[j]>a[j+1]:
            t=a[j]
            a[j]=a[j+1]
            a[j+1]=t
            f=True
        

    if f==False:
        break

for i in a:
    print(i)

print("-------OUTPUT----------")
print("Counter:" ,c)




# it is use for dictionary pattern that is lexicography
# it is order of n square in asc and desc
# it is order of n in average case 
```
