```python
#without parameter and without return

def show1():
    print("Hello")

```

```python
show1()
```

```python
#with parameter and not return

def show2(n):
    print("Hello")

show2(1)
```

```python
#without parameter and return value

def price():
    return 1000

price()
```

```python
#with parameter and return value

def sum(n,b):
    return n+b
```

```python
print(sum(1,2)+10)
```
