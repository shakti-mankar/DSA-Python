```python
def selectionsort(nums):
    n = len(nums)

    for i in range(n):
        c=i
        for j in range(i+1,n):
            if nums[c]>nums[j]:
                
                c=j

        if c != i:
            t=nums[c]
            nums[c]=nums[i]
            nums[i]=t

    return nums
    
```

```python
nums = [5,3,1,8,9]
```

```python
print(selectionsort(nums))
```
