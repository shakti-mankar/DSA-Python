```python
def insertionsort(nums):
    n = len(nums)

    print(n)

    for i in range(1,n):
        j = i
        while j>0 and nums[j]<nums[j-1]:
            t=nums[j]
            nums[j]=nums[j-1]
            nums[j-1]=t
            j=j-1
    return nums
    
```

```python
nums = [5, 3, 4, 1, 2]

print(insertionsort(nums))
```
