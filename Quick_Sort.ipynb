```python
def part(nums,l,r):
    k=nums(r)
    start=l

    for i in range((l,r+1)):
        if nums[i]<=k:
            t=nums[i]
            nums[i]=nums[start]
            nums[start]=t
            start=start+1

    return start-1
```

```python
def qksort(nums,l,r):
    if l>=r:
        return

    p = part(nums,l,r)
    qksort(nums,l,p-1)
    qksort(nums,p+1,r)
```
