```python
# HEAD RECURSION

def rec(s,n):
    if s==n:
        return
    print(s,end=" ")
    return rec(s+1,n)

rec(1,11)     
```

```python
# TAIL RECURSION

def rec(s,n):
    if s==n:
        return
        
    rec(s+1,n)
    print(s,end=" ")

rec(1,11)
```

```python

# FACTORIAL

num = int(input("Enter a number:"))

factorial = 1

for i in range(1,num+1):
    factorial = factorial * i

print("Factorial of " , num , "is" , factorial)
```

```python
# FACTORIAL RECURSION

def Fact(n):
    if n==1:
        return n

    return n*Fact(n-1)

Fact(8)
```
