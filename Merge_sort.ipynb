```python
def merges(nums, l, m, r):
    a = []
    b = []

    for x in range(l, m + 1):
        a.append(nums[x])

    for y in range(m + 1, r + 1):
        b.append(nums[y])

    i, j, k = 0, 0, l

    while k <= r:

        if j == len(b):
            nums[k] = a[i]
            i += 1
            k += 1

        elif i == len(a):
            nums[k] = b[j]      # Fixed
            j += 1
            k += 1

        elif a[i] < b[j]:
            nums[k] = a[i]
            i += 1
            k += 1              # Fixed

        else:
            nums[k] = b[j]
            j += 1
            k += 1
```

```python
def ms(nums,l,r):
    if l>=r:
        return
    
    mid = (l+r) //2 
    ms(nums,l,mid)
    ms(nums,mid+1,r)
    merges(nums,l,mid,r)
    
```

```python
def mgsort(nums):
    n=len(nums)
    ms(nums,0,n-1)
    return nums
```
